# Image Classification Task: Fashion MNIST Dataset

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report,confusion_matrix
import tensorflow as tf 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Dataset url: https://www.kaggle.com/zalando-research/fashionmnist

## Loading Fashion MNIST dataset from Keras

```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.fashion_mnist.load_data()
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

## Creating Test, Train and Validation Datasets

```python
# Reshaping dataset

X_train = X_train.astype(np.float32).reshape(-1, 28*28) / 255.0
X_test = X_test.astype(np.float32).reshape(-1, 28*28) / 255.0
y_train = y_train.astype(np.int32)
y_test = y_test.astype(np.int32)
X_valid, X_train = X_train[:5000], X_train[5000:]
y_valid, y_train = y_train[:5000], y_train[5000:]
```

```python
image = X_train[60,:].reshape((28,28))
plt.imshow(image)
plt.show()
```

## Creating and Training Model: DNN Classifier 

```python
feature_cols = [tf.feature_column.numeric_column("X", shape=[28 * 28])]

dnn_clf = tf.estimator.DNNClassifier(hidden_units=[300,150], n_classes=10,
                                     feature_columns=feature_cols)

input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_train}, y=y_train, num_epochs=40, batch_size=50, shuffle=True)

dnn_clf.train(input_fn=input_fn)
```

## Evaluating validation set results

```python
valid_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_valid}, y=y_valid, shuffle=False)

valid_eval_results = dnn_clf.evaluate(input_fn=valid_input_fn)
```

```python
valid_eval_results
```

## Evaluating model on test data

```python
test_input_fn = tf.estimator.inputs.numpy_input_fn(
    x={"X": X_test}, y=y_test, shuffle=False)

test_eval_results = dnn_clf.evaluate(input_fn=test_input_fn)
```

```python
test_eval_results
```

## Actual and predicted values

```python
y_pred_iter = dnn_clf.predict(input_fn=test_input_fn)
y_pred = list(y_pred_iter)
print('Prediction: ', y_pred[0], '\nLabel: ', y_test[0])
```

## Sensitivity Analysis for Number of Neurons in Hidden Layer 1 and 2

```python
neurons = [100, 200, 300, 400, 500]
accuracy = []
loss = []

for neuron in neurons:

    dnn_clf = tf.estimator.DNNClassifier(hidden_units=[neuron,neuron], n_classes=10,
                                     feature_columns=feature_cols)

    input_fn = tf.estimator.inputs.numpy_input_fn(
        x={"X": X_train}, y=y_train, num_epochs=20, batch_size=30, shuffle=True)

    dnn_clf.train(input_fn=input_fn)
    
    results = dnn_clf.evaluate(input_fn=input_fn)
    
    accuracy.append(results['accuracy'])
    
    loss.append(results['loss'])
```

```python
%matplotlib inline

plt.plot(neurons, accuracy, 'r-', label='Training Accuracy')

plt.plot(neurons, loss, 'b-', label='Training Loss')

plt.title('Training Accuracy and Loss vs Number of Neurons')

plt.legend()

plt.xlabel('Number of Neurons in Hidden Layer 1 and 2')

plt.ylabel('Training Accuracy and Loss')
```
